```python
# reload
Financial_Hardships = pd.read_csv("/Users/deannalakshman/Desktop/IIT/YEAR 2/SEM 2/DSPLC/Individual technical/Raw_Datasets/Finacial hardships.csv")

# drop 1985-1999 year columns
cols_to_drop = [str(year) for year in range(1985, 2000)]
Financial_Hardships.drop(columns=cols_to_drop, inplace=True, errors='ignore')

# keep only the columns you need
cols_to_keep = ['REF_AREA_LABEL', 'FINANCIAL_HARDSHIP_LABEL', 'URBANIZATION_LABEL', 
                'IC_QUINTILE_LABEL', 'INDICATOR_LABEL'] + \
               [col for col in Financial_Hardships.columns if col.isdigit()]

Financial_Hardships = Financial_Hardships[cols_to_keep]

# remove regional aggregates
regions_to_remove = [
    'East Asia & Pacific', 'Europe & Central Asia', 'World',
    'High income', 'Latin America & Caribbean', 'Low income',
    'Lower middle income', 'Middle East, North Africa, Afghanistan & Pakistan',
    'North America', 'South Asia', 'Sub-Saharan Africa', 'Upper middle income'
]
Financial_Hardships = Financial_Hardships[~Financial_Hardships['REF_AREA_LABEL'].isin(regions_to_remove)]

print(Financial_Hardships.shape)
Financial_Hardships.head()
```

```python
Financial_Hardships.drop(columns=['2024'], inplace=True, errors='ignore')
```

```python
Financial_Hardships.drop_duplicates(inplace=True)
```

```python
fh_label_map = {
    'All financial hardship': 'All',
    '1: Facing large (but non-impoverishing) out-of-pocket health expenditure': 'Large OOP',
    '2: Facing impoverishing out-of-pocket health expenditure, based on the societal poverty line': 'Impoverishing',
    '2.1: Pushed into poverty (based on societal poverty line) due to out-of-pocket health expenditure': 'Pushed into poverty',
    '2.2: Further impoverished due to out-of-pocket health expenditure, based on the societal poverty line': 'Further impoverished'
}
Financial_Hardships['FINANCIAL_HARDSHIP_LABEL'] = Financial_Hardships['FINANCIAL_HARDSHIP_LABEL'].map(fh_label_map)
```

```python
FH_total = Financial_Hardships[
    (Financial_Hardships['URBANIZATION_LABEL'] == 'Total') &
    (Financial_Hardships['IC_QUINTILE_LABEL'] == 'All') &
    (Financial_Hardships['FINANCIAL_HARDSHIP_LABEL'] == 'All')
]

print(FH_total.shape)
print(FH_total['REF_AREA_LABEL'].nunique())
```

```python
HC_total = Health_Coverage[Health_Coverage['INDEX_COMP_LABEL'] == 'Full index']
print(HC_total.shape)
```

```python
print(HC_total.columns.tolist())
print(HC_total.shape)
```

```python
print(Financial_Hardships.columns.tolist())
print(Financial_Hardships.shape)
```

```python
year_cols_hc = [col for col in HC_total.columns if col.isdigit()]

HC_long = HC_total.melt(
    id_vars=['REF_AREA_LABEL', 'INDEX_COMP_LABEL'],
    value_vars=year_cols_hc,
    var_name='year',
    value_name='hc_coverage'
)

HC_long['year'] = HC_long['year'].astype(int)
print(HC_long.shape)
```

```python
year_cols_fh = [col for col in Financial_Hardships.columns if col.isdigit()]

FH_long = Financial_Hardships.melt(
    id_vars=['REF_AREA_LABEL', 'FINANCIAL_HARDSHIP_LABEL', 'URBANIZATION_LABEL', 'IC_QUINTILE_LABEL'],
    value_vars=year_cols_fh,
    var_name='year',
    value_name='financial_hardship'
)

FH_long['year'] = FH_long['year'].astype(int)
print(FH_long.shape)
```

```python
merged = pd.merge(
    HC_long,
    FH_long,
    on=['REF_AREA_LABEL', 'year'],
    how='inner'
)

print(merged.shape)
print(merged['REF_AREA_LABEL'].nunique())
merged.head()
```

```python
merged = merged.drop(columns=['INDEX_COMP_LABEL'])
```

merged.to_csv("Cleaned_Datasets/Health_Coverage_Financial_Hardship.csv", index=False)
